```python
import pandas as pd
import numpy as np

from bokeh.plotting import figure, output_file, output_notebook, show
from bokeh.layouts import gridplot, row, column, widgetbox

output_notebook()

TOOLS = 'pan, box_zoom, reset, box_select, lasso_select, wheel_zoom'
plot = figure(title='Random Graph', x_axis_label='X Label', y_axis_label='Y Label', 
              plot_width=300, plot_height=300, 
              x_range=(0, 10), y_range=(0,10),
              tools=TOOLS)

plot.circle(x=[1, 2, 3, 4], y=[2, 4, 6, 1], size=10, legend=False)


plot2 = figure(title='Random Graph 2', plot_width=300, plot_height=300,
               x_range=plot.x_range, y_range=(0,10))
plot2.circle(x=[1, 2, 3, 4], y=[2, 5, 1, 9], size=10, legend=False)

p = gridplot([[plot, plot2]])

show(p)

# # put the results in a row
# show(row(s1, s2, s3))

# # put the results in a column and show
# show(column(s1, s2, s3))

# # make a grid
# grid = gridplot([[p1, p2], [None, p3]])
```

```python
file = '/Users/pmcorrea/Notebooks/country-pops.csv'

countries = pd.read_csv(file, nrows=10)
countries_df = pd.DataFrame(countries)

countries_name = countries_df['Country_English'][0:10]
countries_pop = countries_df['Population'][0:10]
countries_array = np.array(countries_name)
countries_pop_array = np.array(countries_pop)

output_notebook()

p = figure(plot_height=1000, title="Population", toolbar_location=None, tools="")
p.vbar(x=countries_array, top=countries_pop_array)

show(p)
```
