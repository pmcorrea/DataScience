```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
height = np.round(np.random.exponential(3, 100000), 2)
plt.hist(height, bins=100)
plt.title('Exponential Distribution')
plt.ylabel('No. of Samples')
plt.show()
```

```python
array = []
x = 0

while x < 100000:
    sample_mean = np.mean(np.random.choice(height, size=1000, replace=True))
    array = np.append(array, sample_mean)
    x = x + 1
```

```python
plt.hist(array, bins=100)
plt.title('Distribution of Sampled Means')
plt.ylabel('No. of Samples')
plt.xlabel('Mean of Samples')
plt.show()
```
